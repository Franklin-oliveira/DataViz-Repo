## Ajustes no dataset

Vamos passar os datasets para o formato *long*.

```python
# antes...
mar.head(3)
```

```python
# casamentos
mar = mar.set_index('Country')
mar = mar.stack().reset_index()
mar.columns = ['Country', 'Year', 'marriage_rate']

# depois...
mar.head()
```

```python
# divórcios
div = div.set_index('Country')
div = div.stack().reset_index()
div.columns = ['Country', 'Year', 'divorce_rate']

div.head()
```

#### Agregando os datasets

```python
df = pd.merge(mar, div, how='inner', on=['Country', 'Year'])
```

```python
df.head(2)
```

<br>

### Coletando idade média

Vamos coletar a idade média na qual os homens e mulheres casaram e adicionar ao dataframe principal.

```python
mean = excel.parse('MeanAgeFirstMarriage', sep=';', encoding='utf-8-sig', header=3, nrows=82)
mean.replace(to_replace = '..', value = np.NAN, inplace = True) # trocando '..' por NaN
mean.drop('Note', axis=1, inplace=True)

mean.head(3)
```

#### pequenos ajustes...

```python
# preenchendo os valores abaixo de cada país com o nome do país
for country in mean[~mean['Country'].isna()]['Country'].unique():
    i = mean[mean['Country']==country].index[0]
    mean.iloc[i+1,0] = mean.iloc[i, 0]
```

```python
mean.set_index(['Country', 'Gender'], inplace=True)
mean = mean.stack().reset_index()
mean = mean.rename(columns={'level_2':'Year', 0:'mean_age'})
```

```python
mean.head()
```

#### coletando idades 

```python
male_age = mean[mean['Gender']=='Male'].copy()
female_age = mean[mean['Gender']=='Female'].copy()

# renomeando colunas
male_age.rename(columns={'mean_age':'male_mean_age'}, inplace=True)
female_age.rename(columns={'mean_age':'female_mean_age'}, inplace=True)
```

#### agregando novas informações...

```python
df = pd.merge(df, male_age.drop('Gender', axis=1), how='inner', on=['Country', 'Year'])
```

```python
df = pd.merge(df, female_age.drop('Gender', axis=1), how='inner', on=['Country', 'Year'])
```

```python
# criando coluna com idade média do casal
df['couple_mean_age'] = df[['male_mean_age', 'female_mean_age']].apply(np.mean, axis=1)
```

**Criando coluna com os nomes dos continentes**

```python
continents = {
    'NA': 'North America',
    'SA': 'South America', 
    'AS': 'Asia',
    'OC': 'Australia',
    'AF': 'Africa',
    'EU': 'Europe'
}
```

```python
def continente(pais):
    if pais == 'Korea':
        return 'AS'
    else:
        country_code = pc.country_name_to_country_alpha2(pais, cn_name_format="default")
        continent_name = pc.country_alpha2_to_continent_code(country_code)
    return continents[continent_name]
```

```python
df['Continent'] = df['Country'].apply(lambda x:continente(x))
```

## Tentativa de gráfico "gapminder-like"

```python
mar['Continent'] = mar['Country'].apply(lambda x:continente(x))
```

```python
teste = mar.groupby(['Continent', 'Year']).mean().reset_index()
```

```python
# Média das taxas de casamento por continente
alt.Chart(teste).mark_line(filled=False).encode(
    y= alt.Y('marriage_rate'),
    x= alt.X('Year', type='nominal'),
    #size= alt.Size('couple_mean_age', scale= alt.Scale(range=[10, 1000], zero=False)),
    color= alt.Color('Continent', type='nominal'),
    tooltip= alt.Tooltip('Continent'),
    #order= alt.Order('couple_mean_age', sort='descending')
)
```

```python
# gráfico com slider ("dinâmico")
slider = alt.binding_range(min=1990, max=2017, step=1, name='Ano:')

selector = alt.selection_single(name="SelectorName", fields=['Year'],
                                bind=slider)

alt.Chart(df).mark_point(filled=True).encode(
    x= alt.X('marriage_rate', scale= alt.Scale(domain=[0, 15])),
    y= alt.Y('divorce_rate', scale= alt.Scale(domain=[0, 5])),
    size= alt.Size('couple_mean_age', scale= alt.Scale(range=[10, 1000], zero=False)),
    color= alt.Color('Continent', type='nominal'),
    tooltip= alt.Tooltip('Country'),
    order= alt.Order('couple_mean_age', sort='descending')
).properties(
    width=600,
    height=400
).add_selection(
    selector
).transform_filter(
    selector
).interactive()
```

### Mesmo gráfico em escala logarítmica

```python
# escala logaritmica
df['log_marriage_rate'] = df['marriage_rate'].apply(np.log)
df['log_divorce_rate'] = df['divorce_rate'].apply(np.log)
```

```python
# gráfico com slider ("dinâmico") - em escala logarítmica
slider = alt.binding_range(min=1990, max=2017, step=1, name='Ano:')

selector = alt.selection_single(name="SelectorName", fields=['Year'],
                                bind=slider)

alt.Chart(df).mark_point(filled=True).encode(
    x= alt.X('log_marriage_rate', scale= alt.Scale(domain=[0, 3])),
    y= alt.Y('log_divorce_rate', scale= alt.Scale(domain=[-2, 2])),
    size= alt.Size('couple_mean_age', scale= alt.Scale(range=[10, 1000], zero=False)),
    color= alt.Color('Country', type='nominal'),
    tooltip= alt.Tooltip('Country'),
    order= alt.Order('couple_mean_age', sort='descending')
).properties(
    width=600,
    height=400
).add_selection(
    selector
).transform_filter(
    selector
).interactive()
```
